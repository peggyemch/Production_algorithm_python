```python
import numpy as np
import pandas as pd
```

```python
#create the product-times array
prodtime = [['A', 7], ['B', 13], ['C', 2], ['D', 4], ['E', 21], ['F', 6], ['G', 8], ['H', 12], ['I', 17], ['J', 22]]
df_prodtime = pd.DataFrame(prodtime, columns = ['Product', 'Time'], index=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'])
matrix = pd.read_csv('changeover_matrix2.csv')
matrix_df = pd.DataFrame(matrix, columns=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'])
```

```python
matrix_new = matrix.rename(index={0: 'A', 1: 'B', 2: 'C', 3: 'D', 4: 'E', 5: 'F', 6: 'G', 7: 'H', 8: 'I', 9: 'J'})
matrix_new
```

```python
#save the starting list
given = ['B', 'G', 'E', 'A', 'C','I', 'J', 'D', 'H', 'F']
given2 = ['H', 'F', 'G', 'I', 'C','A', 'B', 'D', 'E', 'J']

#find production time
#start with time to produce all elements in the given list
total = 0
for i in range(0, len(given)):
    total = total + int(df_prodtime.loc[given[i], 'Time'])
#add time to swtich between products
for i in range(0, len(given)-1):
    total = total + (int(matrix_new.loc[given[i], given[i+1]]))
print('The given sequence:', given)
print('Total production time:', total, 'units')
```

```python
#if range is even

if (len(given) % 2) == 0:
    while(True):
        swapped = False
        for i in range(1, len(given)-2, 2):
            middleleft = int(matrix_new.loc[given[i], given[i-1]])
            rightmiddle = int(matrix_new.loc[given[i+1], given[i]])
            middleright = int(matrix_new.loc[given[i], given[i+1]])
            leftmiddle = int(matrix_new.loc[given[i-1], given[i]])
            leftright = int(matrix_new.loc[given[i-1], given[i+1]])
            original = leftmiddle + middleright
            leftswap = middleleft + leftright
            rightswap = rightmiddle + leftright
            if original <= leftswap and original <= rightswap:
                break
            elif rightswap <= original and rightswap <= leftswap:
                temp = given[i]
                given[i] = given[i+1]
                given[i+1] = temp
                swapped = True
            elif leftswap <= original and leftswap <= rightswap:
                temp = given[i]
                given[i] = given[i-1]
                given[i-1] = temp
                swapped = True
        if swapped == False:
            break
        #once the loop is done, check the last and penultimate elements
        if int(matrix_new.loc[given[len(given)-1], given[len(given)-2]]) < int(matrix_new.loc[given[len(given)-2], given[len(given)-1]]):
                temp = given[len(given)-1]
                given[len(given)-1] = given[len(given)-2]
                given[len(given)-2] = temp
                swapped = True
                
#if range is odd                
elif (len(given) % 2) != 0:
    while(True):
        swapped = False
        for i in range(1, len(given)-1, 2):
            middleleft = int(matrix_new.loc[given[i], given[i-1]])
            rightmiddle = int(matrix_new.loc[given[i+1], given[i]])
            middleright = int(matrix_new.loc[given[i], given[i+1]])
            leftmiddle = int(matrix_new.loc[given[i-1], given[i]])
            leftright = int(matrix_new.loc[given[i-1], given[i+1]])
            original = leftmiddle + middleright
            leftswap = middleleft + leftright
            rightswap = rightmiddle + leftright
            if original <= leftswap and original <= rightswap:
                break
            elif rightswap <= original and rightswap <= leftswap:
                temp = given[i]
                given[i] = given[i+1]
                given[i+1] = temp
                swapped = True
            elif leftswap <= original and leftswap <= rightswap:
                temp = given[i]
                given[i] = given[i-1]
                given[i-1] = temp
                swapped = True
        if swapped == False:
            break

total = 0
#add time to produce each product
for i in range(0, len(given)):
    total = total + int(df_prodtime.loc[given[i], 'Time'])
total
#add time to swtich between products
for i in range(0, len(given)-1):
    total = total + (int(matrix_new.loc[given[i], given[i+1]]))

print("Best sequence:", given)
print('Total production time:', total, 'units')
```
